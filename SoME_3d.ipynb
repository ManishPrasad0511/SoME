```python
from manim import *
import numpy as np
import itertools
```

```python
a = 8
b = 6
c = 4
sides = [a,b,c]
N = 2
p1 = [-2,1,1]
p2 = [2,-2,-1]
d = 3
```

```python
signs  = list(itertools.product([1, -1], repeat=d))
final_pts = []
for i in signs:
  shifted_p2 = p2 + 2*(N-1)*np.multiply(i,sides)
  #print(shifted_p2)
  for j in range(d):
    final_shift = np.zeros(d)
    final_shift[j] = np.multiply(i,2*(sides[j]/2 - i[j]*(p2[j])))[j]
    #print(final_shift)
    final_pts.append(shifted_p2 + final_shift)
print(final_pts)
len(final_pts)
```

```python
final_coords_all = []

#for p in final_pts:
p = final_pts[0]

x_inters = abs(p[0])//(sides[0]/2) - abs(p[0])//sides[0]
y_inters = abs(p[1])//(sides[1]/2) - abs(p[1])//sides[1]
z_inters = abs(p[2])//(sides[2]/2) - abs(p[2])//sides[2]

x_walls = []
y_walls = []
z_walls = []
if p[0]>=0:
  x_start = sides[0]/2
else:
  x_start = -sides[0]/2
if p[1]>0:
  y_start = sides[1]/2
else:
  y_start = -sides[1]/2
if p[2]>0:
  z_start = sides[2]/2
else:
  z_start = -sides[2]/2

dirn = p-p1
dirn
for i in np.arange(x_inters):
  t = (x_start - p1[0])/dirn[0]
  y = p1[1] + t*dirn[1]
  z = p1[2] + t*dirn[2]
  x_walls.append([x_start,y,z])
  x_start = x_start + np.sign(p[0])*sides[0]

for i in np.arange(y_inters):
  t = (y_start - p1[1])/dirn[1]
  x = p1[0] + t*dirn[0]
  z = p1[2] + t*dirn[2]
  y_walls.append([x,y_start,z])
  y_start = y_start + np.sign(p[1])*sides[1]

for i in np.arange(z_inters):
  t = (z_start - p1[2])/dirn[2]
  y = p1[1] + t*dirn[1]
  x = p1[0] + t*dirn[0]
  z_walls.append([x,y,z_start])
  z_start = z_start + np.sign(p[2])*sides[2]

all_inters = np.array(x_walls+y_walls+z_walls)
indexlist =np.argsort(np.linalg.norm(all_inters,axis=1))
all_inters = all_inters[indexlist]
final_coord_out = all_inters                #useful for outside animation
all_inters
```

```python
grid_dim = []
grid_coords = []
final_coords = []
for i in range(len(all_inters)):
  all_inters[i] += np.array(sides)/2
  grid_dim.append([all_inters[i][0]%sides[0] , all_inters[i][1]%sides[1] , all_inters[i][2]%sides[2]])
  grid_coords.append([all_inters[i][0]//sides[0] , all_inters[i][1]//sides[1] , all_inters[i][2]//sides[2]])
  single_final_coord=np.zeros(d)
  for j in range(d):
    if grid_coords[i][j]%2 == 0:
      single_final_coord[j] = (grid_dim[i][j]) - sides[j]/2
    else:
      single_final_coord[j] = sides[j] - (grid_dim[i][j]) - sides[j]/2
  
  final_coords.append(single_final_coord)

final_coords_all.append(final_coords)
```

```python
for i in range(len(final_coords_all)):
  for j in range(len(final_coords_all[i])-1,-1,-1):
    if j==len(final_coords_all[i])-1:
      final_coords_all[i].append( np.array(p2) - final_coords_all[i][j])
    if j==0:
      final_coords_all[i][j] -= np.array(p1)
    else:
      final_coords_all[i][j] -=  final_coords_all[i][j-1]
```

```python
%%manim cuboid_mirror

class cuboid_mirror(ThreeDScene):
    def construct(self):
      self.set_camera_orientation(phi=80*DEGREES, theta=-60*DEGREES, distance=14)
      self.begin_ambient_camera_rotation()
      cuboid = Prism(dimensions=[sides[0], sides[1], sides[2]]).set_fill(BLUE, opacity=0.2)

      dot = Sphere(center=np.array([p1[0], p1[1], p1[2]]) , color=BLUE, radius=0.08)
      d6  = Sphere(center=np.array([p2[0], p2[1], p2[2]]) , color=BLUE, radius=0.08)
      
      
      def update_path(path):
            previous_path = path.copy()
            previous_path.add_points_as_corners([dot.get_center()])
            path.become(previous_path)
      
      self.add(cuboid)
      self.wait()
      self.add(dot,d6)
      self.wait()
      
      #self.move_camera(0.8*np.pi/2, -0.45*np.pi)


      #for i in range(2): 
      for i in range(len(final_coords_all)):
        dot = Sphere(center=np.array([p1[0], p1[1], p1[2]]) , color=BLUE, radius=0.08)

        path = VMobject()     
        path.set_points_as_corners([dot.get_center(), dot.get_center()])
        path.add_updater(update_path)
        
        self.add(path, dot)
        dot.generate_target()
        for j in range(len(final_coords_all[i])):
          dot.target.shift(final_coords_all[i][j][0]*RIGHT , final_coords_all[i][j][1]*UP , final_coords_all[i][j][2]*OUT)
          self.play(MoveToTarget(dot))

        self.play(FadeOut(dot,path))
        self.remove(dot,path)
```

```python
%%manim three_d_explanation
class three_d_explanation(ThreeDScene):
    def construct(self):
      self.set_camera_orientation(phi=80*DEGREES, theta=-60*DEGREES, distance=20)
      self.begin_ambient_camera_rotation()
      scale = 10

      cuboid = Prism(dimensions=[sides[0], sides[1], sides[2]]).set_fill(BLUE, opacity=0.2)

      dot = Sphere(center=np.array([p1[0], p1[1], p1[2]])/scale , color=BLUE, radius=0.08)
      d6  = Sphere(center=np.array([p2[0], p2[1], p2[2]])/scale , color=BLUE, radius=0.08)
      
      all_rects_vgroup = VGroup()
       
      for i in np.arange(-3*sides[0]/scale ,3*sides[0]/scale+0.1 , sides[0]/scale):
        row_rects = VGroup()
        for j in np.arange(-3*sides[1]/scale ,3*sides[1]/scale+0.1 , sides[1]/scale):
          col_rects = VGroup()
          for k in np.arange(-3*sides[2]/scale ,3*sides[2]/scale+0.1 , sides[2]/scale):
            new_cuboid = Prism(dimensions=np.array([sides[0], sides[1], sides[2]])/scale).set_fill(BLUE, opacity=0.1)
            new_cuboid.shift(i*RIGHT,j*UP,k*OUT)
            new_cuboid.set_stroke(color=WHITE, width=0.4)

            col_rects.add(new_cuboid)
          row_rects.add(col_rects)
        all_rects_vgroup.add(row_rects)
      
      self.add(dot,d6)
      self.wait()
      self.add(all_rects_vgroup[3][3][3])
      self.wait()
      self.add(all_rects_vgroup)
      all_rects_vgroup[3][3][3].set_fill(PINK, opacity=1)

      arrow_labs = list(itertools.product([2, -2], repeat=d))
      all_arrows = VGroup()
      for i in arrow_labs:
        all_arrows.add(Vector(direction=np.array([i[0]*sides[0]/scale, i[1]*sides[1]/scale, i[2]*sides[2]/scale]),
                              buff=0,
                              color=RED))
      self.play(FadeIn(all_arrows))
      
      signs = list(itertools.product([1, -1], repeat=d))
      box_labels = [(i + np.array([3,3,3])) for i in list(itertools.product([2, -2], repeat=d))]    
      for i in range(len(box_labels)):
        all_rects_vgroup[box_labels[i][0]][box_labels[i][1]][box_labels[i][2]].set_fill(RED, opacity=1)
      self.wait()
      for i in range(len(box_labels)):
        all_rects_vgroup[box_labels[i][0] + signs[i][0]][box_labels[i][1]][box_labels[i][2]].set_fill(YELLOW, opacity=0.5)
        all_rects_vgroup[box_labels[i][0]][box_labels[i][1] + signs[i][1]][box_labels[i][2]].set_fill(YELLOW, opacity=0.5)
        all_rects_vgroup[box_labels[i][0]][box_labels[i][1]][box_labels[i][2] + signs[i][2]].set_fill(YELLOW, opacity=0.5)


      

```
